## Algerian Forest Fires Dataset 
Data Set Information:

The dataset includes 244 instances that regroup a data of two regions of Algeria, namely the Bejaia region located in the northeast of Algeria and the Sidi Bel-abbes region located in the northwest of Algeria.

122 instances for each region.

The period from June 2012 to September 2012.

The 244 instances have been classified into fire(138 classes) and not fire (106 classes) classes.

Attribute Information:

1. Date : (DD/MM/YYYY) Day, month ('June' to 'September'), year (2012)
Weather data observations
2. Temp : temperature noon (temperature max) in Celsius degrees: 22 to 42
3. RH : Relative Humidity in %: 21 to 90
4. Ws :Wind speed in km/h: 6 to 29
5. Rain: total day in mm: 0 to 16.8
FWI Components
6. Fine Fuel Moisture Code (FFMC) index from the FWI system: 28.6 to 92.5
7. Duff Moisture Code (DMC) index from the FWI system: 1.1 to 65.9
8. Drought Code (DC) index from the FWI system: 7 to 220.4
9. Initial Spread Index (ISI) index from the FWI system: 0 to 18.5
10. Buildup Index (BUI) index from the FWI system: 1.1 to 68
11. Fire Weather Index (FWI) Index: 0 to 31.1
12. Classes: two classes, namely Fire and not Fire

## Data Cleaning

```python
#Checking for missing values
dataset[dataset.isnull().any(axis=1)]
```

The dataset converted into two sets based on Region from 122th index, we can make a new column based on the Region

1 : "Bejaia Region Dataset" == 1
2 : "Sidi-Bel Abbes Region Dataset" ==0

Add new column with region

```python
dataset.loc[:122,"Region"]=0
dataset.loc[122:,"Region"]=1
df=dataset
```

```python
df.info()
```

```python
#Converting the 'Region' column to integer
df[['Region']]=df[['Region']].astype(int)
```

```python
df.head()
```

```python
#Checking for Null values
df.isnull().sum()
```

```python
## Removing the null values
df=df.dropna().reset_index(drop=True)
```

```python
df.head()
```

```python
## Removing the null values again
df.isnull().sum()
```

```python
#Checking for the column head of Sidi-Bel Abbes Region Dataset so I can remove it
df.iloc[[122]]
```

```python
#Removing the 122nd row
df=df.drop(122).reset_index(drop=True)
```

```python
#Checking row 122 again to see if the column head of Sidi-Bel Abbes Region Dataset has been remove
df.iloc[[122]]
```

```python
#Checking the column names
df.columns
```

```python
#Removing spaces in columns names
df.columns=df.columns.str.strip()
df.columns
```

```python
df.info()
```

#### Changing the some columns data type and converting it to integer data type

```python
df.columns
```

```python
#Converting the columns to integer data type
df[['month','day','year','Temperature','RH','Ws']]=df[['month','day','year','Temperature','RH','Ws']].astype(int)
```

```python
df.info()
```

#### Changing the other columns to float data datatype


```python
#Using list comprehension to create a list of column names from my DataFrame
objects=[features for features in df.columns if df[features].dtypes=='O']
```

```python
#Using a loop to iterate over my DataFrame to convert the datatypes that are Object to float except 'Classes' column
for i in objects:
    if i!='Classes':
        df[i]=df[i].astype(float)
```

##  Exploratory Data Analysis

```python
#Droping unwanted columns and making a copy of the new dataset
df_copy=df.drop(['day','month','year'],axis=1)
```

```python
df_copy.head()
```

```python
#Checking categories in Classes Column
df_copy['Classes'].value_counts()
```

```python
#Encoding the categories in classes to 0 and 1
df_copy['Classes']=np.where(df_copy['Classes'].str.contains('not fire'),0,1)
```

```python
df_copy.head()
```

```python
#Bottom 5 rows
df_copy.tail()
```

```python
#Count of unique values in 'Classes' column
df_copy['Classes'].value_counts()
```

```python

#Determining the number of columns for my subplot
num_columns = 3
num_features = len(df_copy.columns)
num_rows = (num_features + num_columns - 1) // num_columns  # Calculate needed rows

# Creating the subplots
plt.figure(figsize=(15, num_rows * 5))  # Adjusting figure size

for i, column in enumerate(df_copy.columns):
    plt.subplot(num_rows, num_columns, i + 1)  # Creating a subplot for each column
    plt.hist(df_copy[column], bins=10, alpha=0.7, color='g', edgecolor='black')
    plt.title(f'{column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.grid(axis='y', alpha=0.75)

plt.tight_layout()  # Adjusting layout
plt.show()  
```

```python
#Percentage for Pie Chart
percentage=df_copy['Classes'].value_counts(normalize=True)*100
```

```python
percentage
```

```python
#Plotting piechart
classlabels=["Fire","Not Fire"]
plt.figure(figsize=(12,7))
plt.pie(percentage,labels=classlabels,autopct='%1.1f%%')
plt.title("Pie Chart of Classes")
plt.show()
```
